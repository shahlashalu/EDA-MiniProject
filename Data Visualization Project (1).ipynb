# **CLEANING**

```python
df.Rating.value_counts()
```

```python
# Changing "Previous" to Nan before replacing with mode
df['Rating'] = df['Rating'].replace("Previous", np.nan)
```

```python
# Replacing Nan in Rating with Mode
Rating_mode = df['Rating'].mode()[0]
df['Rating'] = df['Rating'].fillna(Rating_mode)
```

```python
df['Brand'].unique()
```

```python
#Capitalizing first character of Brand
df['Brand'] = df['Brand'].str.capitalize()
```

```python
df.head()
```

```python
df['Screen Size']=df['Screen Size'].str.split(' ').str[0].astype(float)
```

```python
df.info()
```

```python
#Handling Rare Categorical feature: Resolution column
resol = pd.DataFrame(df['Resolution'].value_counts()).reset_index()
resol = resol[(resol['count']/df.shape[0])*100.0 > 1.0]
```

```python
resol_cat = resol['Resolution'].tolist()
resol_cat
```

```python
#replacing rare Categories by 'Other' in Resolution Column
df['Resolution'] = np.where(df['Resolution'].isin(resol_cat), df['Resolution'], 'Other')
```

```python
df.Resolution.value_counts()
```

```python
df.head()
```

```python
df['Aspect Ratio'].unique()
```

```python
df['Aspect Ratio'].value_counts()
```

```python
# handle rare Categories in column Aspect Ratio:
ratios = pd.DataFrame(df['Aspect Ratio'].value_counts()).reset_index()
ratios = ratios[(ratios['count']/df.shape[0])*100.0 > 1.0]
ratios_cat = ratios['Aspect Ratio'].tolist()
```

```python
#replacing rare Categories by 'Other' in Resolution Column
df['Aspect Ratio'] = np.where(df['Aspect Ratio'].isin(ratios_cat), df['Aspect Ratio'], 'Unknown')
```

```python
df['Aspect Ratio'].value_counts()
```

```python
df.head()
```

# **VISUALISATION**

```python
sales_count = df.Title.value_counts().reset_index()
sales_count['Title'] = sales_count['Title'].str[:20]
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
sns.set_palette('mako')
```

```python
# Best-Sellers : Monitors

sns.barplot(data= sales_count[:10] , y='Title' , x='count', palette = 'magma' )

# plt.xticks(rotation=90)
plt.xlabel('Units Sold')
plt.ylabel('Product')
plt.title('Top 10 Best Sellers: Monitors')
plt.show()
```

```python
#best selling Brands
brands_count = df.Brand.value_counts().reset_index()

sns.barplot(data= brands_count[:5] , y='Brand' , x='count' , palette='rocket' )
plt.xlabel('Units Sold')
plt.ylabel('Brand')
plt.title('Top 5 Best Selling Brands')
plt.show()
```

Observation: Acer is the leading Brand in Market



```python
# most popular Screen size

sns.histplot(data= df, x='Screen Size', kde=True, bins=15 )
plt.xlabel('Screen Size (inches)')
plt.ylabel('Frequency')
plt.title('Screen Size Distribution in Market')
plt.show()
```

```python
# pie chart : most popular Screen Resolution

rsln = df['Resolution'].value_counts().reset_index()
labels= rsln['Resolution'].tolist()
counts= rsln['count'].tolist()

palette = sns.color_palette("Set3", n_colors=5)
explode = [0.1,0,0,0,0]

plt.pie(counts, labels=labels, autopct='%1.2f%%',explode=explode, colors=palette, textprops={ 'fontsize': 12})
plt.title('Resolution: Popular Choices')
plt.show()
```

```python
# Popular Aspect Ratios

tempo = df['Aspect Ratio'].value_counts().reset_index()
labels= ['16:9','21:9','Others','16:10']
counts= tempo['count'].tolist()

palette = sns.color_palette("viridis", n_colors=4)
explode=[0.1,0,0,0]

plt.pie(counts, labels=labels,explode=explode, colors=palette)
plt.title('Popular Aspect Ratios')
plt.show()
```

```python
df['Rating'] =df['Rating'].astype(float)
```

```python
# creating column ratingtext (excellent .... very poor)
# Define bins and labels
bins = [0, 2.5, 3.0, 3.5, 4.0, 4.5, 4.9, 5.0]
labels = ['Terrible', 'Poor', 'Below Average', 'Average', 'Good', 'Very Good', 'Excellent']

# Create the ratingtext column
df['ratingtext'] = pd.cut(df['Rating'], bins=bins, labels=labels, right=True)
```

```python
sns.histplot(data =df, x='Rating', hue='ratingtext',bins=10)
plt.title('Ratings Distribution')
plt.show()
```

```python
#rating text counts
plt.figure(figsize=(8,3))
sns.countplot(data=df , y='ratingtext',palette='Set2')
plt.ylabel(None)
plt.show()
```

**Investigating the Relationship Between Price and Rating**

```python
# Visualize the relationship, along with a regression line to observe trends.

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Price', y='Rating', scatter_kws={'alpha':0.5} )
plt.title('Price vs Rating with Regression Line')
plt.xlabel('Price (USD)')
plt.ylabel('Rating')
plt.show()
```

```python
# Calculate the correlation coefficient to quantify the relationship

correlation = df['Price'].corr(df['Rating'])
print(f'Correlation between Price and Rating: {correlation:.2f}')
```

```python
#  Group prices into bins and analyze rating distributions within those bins

bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, np.inf]
labels = ['0-100', '100-200', '200-300', '300-400', '400-500', '500-600', '600-700', '700-800', '800-900', '900-1000', '1000+']
df['Price Range'] = pd.cut(df['Price'], bins=bins, labels=labels)

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Price Range', y='Rating')
plt.title('Rating Distribution by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
#correlation in Numerical Features: Size , Rating, Price

plt.figure(figsize=(8,4))

# Check the data type of 'Price' column
print(df['Price'].dtype)

# If 'Price' column is of object type, then remove commas and convert to float
if df['Price'].dtype == 'object':
    df['Price'] = df['Price'].str.replace(',', '').astype(float)

corr_matrix=df[['Screen Size','Rating','Price']].corr()
sns.heatmap(corr_matrix)
plt.title('Correlation: Size , Rating, Price')
plt.show()
```

```python
# analysing price
plt.figure(figsize=(9,3))
sns.histplot(data =df , x='Price', bins=50)
plt.xlabel('Price in USD')
plt.ylabel('Frequency')
plt.title('Price(USD) Distribution')
plt.show()
```

```python
# scatter plot for screen size vs price

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Price' , y='Screen Size' , hue='Resolution' )
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel('Price in USD')
plt.ylabel('Screen Size (inches)')
plt.title('ScreenSize vs Price(USD)')
plt.show()
```

```python
# OUTLIER DETECTION using BOXPLOTS
sns.set_palette('flare')
fig, axs = plt.subplots(1, 3, figsize=(15, 8))

sns.boxplot(data=df['Screen Size'], ax=axs[0])
axs[0].set_title('Screen Size(inches)')

sns.boxplot(data=df['Rating'], ax=axs[1])
axs[1].set_title('Rating')

sns.boxplot(data=df['Price'], ax=axs[2])
axs[2].set_title('Price (USD)')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
